### ***We created a data using a dictionary. We converted it to a dataframe using the `pandas.DataFrame()` method. We also customized the indexes of the dataframe.***

```python
data2 = {"Name": ["Ali", "Sara", "Sina", "Dana"],
         "Age": [22, 21, 19, 20],
         "Score": [70, 87, 88, 90],
         "Term": [7, 5, 2, 3],
         "City": ["Esfehan", "Shiraz", "Rasht", "Tehran"]}
df = pd.DataFrame(data2, index=[1, 2, 3, 4])
df
```

### ***Using the `dataframe.head()` method, we can view the first five rows of the data frame. This method displays the first five rows by default, but we can change it to display the number of rows we want by giving a specific number.***

```python
df.head(3)
```

### ***The output of the `dataframe.shape` method is a tuple containing the number of rows and the number of columns.***

```python
df.shape
```

### ***By writing the name of the data frame and writing the name of the desired column in a pair of brackets, we can access the data in that column, the name of that column, and its type.***

```python
df["City"]
```

### ***To access data related to several columns, if there is more than one column, we use two pairs of nested brackets.***

```python
df[["Term", "Age", "Name"]]
```

### ***With the following method, we can see in the output the values ​​that apply to the condition written in a column of data.***

```python
df[df["Age"] > 20]
```

### ***With the following method, we can see in the output the values ​​that apply to the condition written in the data of a column. We can write more than one condition and also pay attention to the order of the operators.***

```python
df[(df["Age"] > 20) & (df["City"] == "Shiraz")]
```

### ***Here we created a 2D array using `numpy.array()`. Using `pandas.DataFrame()` method, we converted the data to a dataframe and customized the row and column names.***

```python
data3 = np.array([[10, 20, 30],
                  [40, 50, 60],
                  [70, 80, 90]])
pd.DataFrame(data3, columns=["A", "B", "C"], index=["Row1", "Row2", "Row3"])
```
