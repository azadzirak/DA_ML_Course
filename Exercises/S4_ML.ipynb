### ***Importing the libraries needed in today's program and exercise. `Numpy`, `Pandas`, `Matplotlib`, `Scikit-Learn`, `arabic-reshaper`, and `bidi`.***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.linear_model import LinearRegression
import arabic_reshaper
from bidi.algorithm import get_display
```

### ***Using the `numpy.aaray()` method, we created a two-dimensional array called `x` that has two pairs of brackets and one-dimensional array called `y` that has one pair of brackets.***

```python
x = np.array([[1], [2], [3], [4]])
y = np.array([20, 30, 40, 50])
```

### ***`Linear Regression` is a Supervised Learning algorithm used for predicting continuous numerical outputs. It works with labeled data to model the straight-line relationship between input variables and the target variable. The algorithm's main goal is to find the best-fit line through the data, enabling accurate predictions for new, unseen data points.***

### ***`LinearRegression()` is a class from the `sklearn` library that is used for the linear regression algorithm, and we have already imported it into the program. We access the methods and attributes inside the LinearRegression() function using the `model` object.***

```python
model = LinearRegression()
```

### ***Fit model describes the relationship between a response variable and one or more predictor variables. General syntax: `model.fit(x_train, y_train)`.***

```python
model.fit(x, y)
```

### ***`Model prediction` refers to the outcome generated by a machine learning model, indicating the likelihood of a specific event or behavior occurring based on input data.***

```python
y_fit = model.predict(x)
x_new = np.array([[5], [6], [7], [8], [9]])
y_pred_new = model.predict(x_new)
```

### ***`numpy.concatenate()` join a sequence of arrays along an existing axis. General syntax: numpy.concatenate((a1, a2, ...), axis=0, out=None, dtype=None, casting="same_kind").***

```python
x_all = np.concatenate((x, x_new))
y_all = np.concatenate((y_fit, y_pred_new))
```

### ***We use the `matplotlib.pyplot` library to draw the desired plot. We use the `plt.scatter()` method to draw the scatter plot section for both the old/original data and the predicted data. We use the `plt.plot()` method to draw the regression line. We use the `plt.title()` method to write the title at the top of the plot. We use the `plt.xlabel()` method to add a label to the x-axis. We use the `plt.ylabel()` method to add a label to the y-axis. We use the `plt.legend()` method to display the plot's legend in the drawn graph. We use the `plt.show()` method to display the drawn plot. When coding in a Jupyter environment, there is no need to include this line as the plot will be displayed regardless. We use the `get_display(arabic_reshaper.reshape("name"))` command to add Persian labels and text to the plot. This method ensures that all characters are displayed correctly and legibly on the plot.***

```python
plt.scatter(x, y, color="blue", marker="*", label=get_display(arabic_reshaper.reshape("داده های نمونه")))
plt.scatter(x_new, y_pred_new, color="teal", marker="*", label=get_display(arabic_reshaper.reshape("نقاط پیش بینی")))
plt.plot(x_all, y_all, color="green", label=get_display(arabic_reshaper.reshape("خط رگرسیون")))
plt.title(get_display(arabic_reshaper.reshape("وضعیت نمره بر اساس ساعت مطالعه")))
plt.xlabel(get_display(arabic_reshaper.reshape("ساعت مطالعه")))
plt.ylabel(get_display(arabic_reshaper.reshape("نمره")))
plt.legend()
plt.show()
```

### ***This code prints the calculated slope (coefficient) and y-intercept of the linear regression model, formatted to two decimal places.***

```python
print(f"line_ship: {model.coef_[0]:.2f} \nwidth_of_origin: {model.intercept_:.2f}")
```

### ***Homework 1: "A marketing company wants to examine the effect of monthly social media advertising costs on the sales rate of its product. The goal is to create a Linear Regression model for predicting future sales based on the advertising budget. Problem Data: Advertising Cost (Million Toman) = 10, 15, 20, 25, 30, 35, 40 and Monthly Sales (Thousand Units) = 110, 135, 160, 180, 205, 220, 245. Task: Using the data above, and by designing the model to receive the advertising cost amount from the user, predict the monthly sales rate."***

```python
Cost = np.array([[10], [15], [20], [25], [30], [35], [40]])
Sales = np.array([110, 135, 160, 180, 205, 220, 245])
model = LinearRegression()
model.fit(Cost, Sales)
```

```python
Sales_fit = model.predict(Cost)
x = int(input("x: "))
Cost_x = np.array([[x]])
Sales_pred = model.predict(Cost_x)
```

```python
Cost_all = np.concatenate((Cost, Cost_x))
Sales_all = np.concatenate((Sales_fit, Sales_pred))
```

```python
plt.title(get_display(arabic_reshaper.reshape("فروش بر اساس بودجه ی تبلیغاتی")))
plt.xlabel(get_display(arabic_reshaper.reshape("هزینه ی تبلیغاتی (میلیون تومان)")))
plt.ylabel(get_display(arabic_reshaper.reshape("فروش ماهانه (هزار واحد)")))
plt.scatter(Cost, Sales, color="red", marker="*", label=get_display(arabic_reshaper.reshape("داده های نمونه")))
plt.scatter(Cost_x, Sales_pred, color="black", marker="*", label=get_display(arabic_reshaper.reshape("نقاط پیش بینی")))
plt.plot(Cost_all, Sales_all, color="gray", label=get_display(arabic_reshaper.reshape("خط رگرسیون")))
plt.legend()
plt.show()
```

### ***Homework 2: "Suppose we want to predict a person's monthly income based on the number of years of work experience. Write an algorithm that performs the predict task. The number of training samples is 5. And 5 inputs are received dynamically from the user for predict."***

```python
Experience = np.array([[1], [2], [3], [4], [5]])
Salary = np.array([20, 28, 32, 40, 47])
model = LinearRegression()
model.fit(Experience, Salary)
```

```python
Salary_fit = model.predict(Experience)
a, b, c, d, e = map(int, input("a, b, c, d, e: ").split())
Experience_x = np.array([[a], [b], [c], [d], [e]])
Salary_pred = model.predict(Experience_x)
```

```python
Experience_all = np.concatenate((Experience, Experience_x))
Salary_all = np.concatenate((Salary_fit, Salary_pred))
```

```python
plt.title(get_display(arabic_reshaper.reshape("حقوق بر حسب سابقه ی کار")))
plt.xlabel(get_display(arabic_reshaper.reshape("سابقه ی کار (سال)")))
plt.ylabel(get_display(arabic_reshaper.reshape("حقوق ماهانه (میلیون تومان)")))
plt.scatter(Experience, Salary, color="brown", marker="*", label=get_display(arabic_reshaper.reshape("داده های نمونه")))
plt.scatter(Experience_x, Salary_pred, color="black", marker="*", label=get_display(arabic_reshaper.reshape("نقاط پیش بینی")))
plt.plot(Experience_all, Salary_all, color="gray", label=get_display(arabic_reshaper.reshape("خط رگرسیون")))
plt.legend()
plt.show()
```
