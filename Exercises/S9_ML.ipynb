### ***Importing required libraries; We start by importing the libraries needed for today's practice.***

### ***Our goal is to work with unsupervised learning algorithms.***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering, KMeans
from sklearn.mixture import GaussianMixture
```

### ***Defining a sample array for hierarchical clustering; We create a simple NumPy array to use in the agglomerative (hierarchical) clustering example.***

```python
x_agg = np.array([[8, 13], [4, 7], [90, 43], [100, 76], [34, 67], [98, 103], [83, 96], [120, 72]])
```

### ***Creating the Agglomerative Clustering model; To use this algorithm, we first create an instance of its class. `n_clusters` specifies the number of groups we want, and `linkage="ward"` tells the algorithm to merge clusters based on minimum variance.***

```python
agg_cluster = AgglomerativeClustering (
    n_clusters=3,
    linkage="ward"
)
```

### ***Fitting the hierarchical clustering model; We fit the model using the array we defined earlier. The model learns how to group the points into clusters.***

```python
agg_cluster.fit(x_agg)
```

### ***Visualizing hierarchical clustering results; We plot the data points and color them using the labels predicted by the agglomerative model.***

```python
plt.figure(figsize=(8,6))
plt.scatter(x_agg[:, 0], x_agg[:, 1], c=agg_cluster.labels_, s=100, edgecolors="k", marker="*", cmap="viridis")
plt.show()
```

### ***Generating data for K-Means (partitioning clustering); For partitioning methods, we use the K-Means algorithm. Here, we generate synthetic data using `make_blobs`.***

```python
x_km, y_km = make_blobs(
    n_samples=300,
    centers=4,
    cluster_std=0.8,
    random_state=42
)
```

### ***Creating the K-Means model; We create a KMeans object with the desired number of clusters.***

```python
kmeans = KMeans(
    n_clusters=4,
    random_state=42
)
```

### ***Training the K-Means model; We fit the K-Means model on the generated dataset so it can learn the cluster structure.***

```python
kmeans.fit(x_km)
```

### ***Predicting cluster labels; Using the trained model, we predict the cluster for each sample.***

```python
y_km_pred = kmeans.predict(x_km)
```

### ***Extracting cluster centers; K-Means calculates the center of each cluster. This step retrieves those centers from the trained model.***

```python
centers = kmeans.cluster_centers_
```

### ***Plotting K-Means results; We draw the scatter plot of data points and highlight the cluster centers on the graph.***

```python
plt.figure(figsize=(10,7))
plt.scatter(x_km[:, 0], x_km[:, 1], c=y_km_pred, s=2000, alpha=0.3, marker="*", label="Cen", cmap="viridis")
plt.scatter(centers[:, 0], centers[:, 1], c="black", s=200, alpha=0.8, marker="*", label="Cen")

plt.legend()
plt.show()
```

### ***Exercise 1 — Customer clustering (K-Means)***

### ***Using K-Means and synthetic data created with `make_blobs`, we cluster customers into 5 groups based on two features (e.g., age and spending). Finally, we visualize the clusters and their centers.***

```python
a_km, b_km = make_blobs(
    n_samples=400,
    centers=5,
    cluster_std=1.5,
    random_state=42
)
```

```python
kmeans = KMeans(
    n_clusters=5,
    random_state=42
)
```

```python
kmeans.fit(a_km)
```

```python
b_km_pred = kmeans.predict(a_km)
```

```python
centers = kmeans.cluster_centers_
```

```python
plt.figure(figsize=(10,7))
plt.scatter(a_km[:, 0], a_km[:, 1], c=b_km_pred, s=2000, alpha=0.3, marker="*", label="Cen", cmap="viridis")
plt.scatter(centers[:, 0], centers[:, 1], c="black", s=200, alpha=0.8, marker="*", label="Cen")

plt.legend()
plt.show()
```

### ***Exercise 2 — Model-based clustering (Gaussian Mixture)***

### ***We build a model-based clustering approach using the Gaussian Mixture Model (GMM). The data contains 3 clusters, and GMM identifies them based on probabilistic distributions.We then plot the assigned clusters along with their estimated means.***

```python
x_gmm, y_gmm = make_blobs(
    n_samples=500,
    centers=3,
    cluster_std=0.6,
    random_state=42
)
```

```python
gmm = GaussianMixture(
    n_components=3,
    random_state=42
)
```

```python
gmm.fit(x_gmm)
```

```python
y_gmm_pred = gmm.predict(x_gmm)
```

```python
centers = gmm.means_
```

```python
plt.figure(figsize=(10,7))
plt.scatter(x_gmm[:, 0], x_gmm[:, 1], c=y_gmm_pred, s=2000, alpha=0.3, marker="*", label="Cen", cmap="viridis")
plt.scatter(centers[:, 0], centers[:, 1], c="black", s=200, alpha=0.8, marker="*", label="Cen")

plt.legend()
plt.show()
```
