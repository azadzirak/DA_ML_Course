### ***To add labels and captions at the top of tables and data frames, we use the `dataframe.style.set_caption("title name")` method.***

```python
data2 = {"Name": ["Ahoora", "Arman", "Mahoor"],
         "Age": [21, 20, 17],
         "City": ["Shiraz", "Tabriz", "Tehran"]}
df2 = pd.DataFrame(data2, index=["a", "b", "c"])
a2 = df2.style.set_caption("Test Student")
a2
```

### ***The loc method is for fetching data from a data frame. General syntax: `DataFrame.loc[rows,cols]`***

```python
df2.loc["b"]
```

### ***To fetch two data from a data frame with the `dataframe.loc` method, we use two pairs of brackets.***

```python
df2.loc[["a", "c"]]
```

### ***We follow the general syntax of the loc method.***

```python
df2.loc[["a", "b"], ["Name", "City"]]
```

### ***According to the general syntax of the loc method and using the condition, we extract the desired data.***

```python
df2.loc[df2["Age"] > 20, ["Name", "Age"]]
```

### ***According to the general syntax of the loc method and using the condition, we can extract the desired data and add a column to the data frame. We note that we do not always use this solution and there is another method specifically for inserting a column into the data frame.***

```python
df2.loc[df2["Age"] < 19, "Status"] = "Minor"
df2
```

### ***We access the methods and attributes inside the `load_iris()` function using the iris_data object. We imported the load_iris function from the sklearn library.  Using the `pandas.DataFrame` method, we convert the stored data, which is in the form of an array, to a data frame. Using the `dataframe.head()` method, we can view the first five rows of the data frame. This method displays the first five rows by default, but we can change it to display the number of rows we want by giving a specific number.***

```python
iris_data = load_iris()
data3 = iris_data.data
feature_name = iris_data.feature_names
data3_with_header = pd.DataFrame(data3, columns=feature_name)
data3_with_header.head(8)
```

### **By default, this function will create a grid of Axes such that each numeric variable in data will by shared across the y-axes across a single row and the x-axes across a single column. The diagonal plots are treated differently: a univariate distribution plot is drawn to show the marginal distribution of the data in each column. `General syntax:` seaborn.pairplot(data, *, hue=None, hue_order=None, palette=None, vars=None, x_vars=None, y_vars=None, kind='scatter', diag_kind='auto', markers=None, height=2.5, aspect=1, corner=False, dropna=False, plot_kws=None, diag_kws=None, grid_kws=None, size=None)***

```python
iris_data = sns.load_dataset("iris")
sns.pairplot(data=iris_data, hue="species", markers="*")
```
