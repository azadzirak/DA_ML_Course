### ***In the code below, we define two 2D arrays and then calculate their matrix product using the `@` operator (at sign). This is the standard and recommended approach in NumPy.***
### ***Note: The commented-out line `# np.matrix(y1) * np.matrix(y2)` achieves matrix multiplication but uses the deprecated np.matrix class. The `@` operator is the modern and preferred equivalent for np.array.***

```python
y1 = np.array([[1, 2]])
y2 = np.array([[2, 3], [1, 4]])
# np.matrix(y1) * np.matrix(y2)
y1 @ y2
```

### ***If we multiply the two 2D arrays using the * operator (e.g., y1 * y2), we get the element-wise product, not the matrix product.***

```python
y1 * y2
```

### ***We import the `Matplotlib` library and the `PyPlot` module using an alias (e.g., `plt`). This library is widely and extensively used for plotting graphs.***

```python
import matplotlib.pyplot as plt
```

### ***We draw a Line Plot (or Line Graph) using the `plt.plot()` method.***

```python
plt.plot([3, 4, 5, 7, 1], [14, 18, 12, 23, 19], color="teal", marker="*", linewidth=2)
```

### ***We draw a Scatter Plot using the `plt.scatter()` method.***

```python
plt.scatter([3, 4, 5, 7, 1], [14, 18, 12, 23, 19], color="green", marker="*", linewidth=2)
```

### ***We draw a Pie Chart using the `plt.pie()` method.***

```python
plt.pie([10, 20, 8, 13], labels=["a", "b", "c", "d"], colors=["green", "teal", "blue", "skyblue"], explode=[0.0, 0.0, 0.0, 0.03])
```
